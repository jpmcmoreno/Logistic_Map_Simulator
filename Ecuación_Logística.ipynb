# 1. La ecuación logística
***

Este modelo esta dado por:

**Ecuación:** $\frac{dP(t)}{dt} = r \cdot P(t) \cdot \left(1- \frac{P(t)}{K} \right)$

* $P(t)$ es la poblacón en tiempo $t$.
* $K$ es la máxima capacidad.
* $r$ es la tasa de crecimiento.

*** 
De forma discreta

**Ecuación:** $x_{n+1} = x_{n} \cdot r \cdot (1 - x_{n})$

* $x_n$ es la poblacón en tiempo $n$. $(0 \leq x_{n} \leq 1)$
* $K$ es la máxima capacidad.
* $r$ es la tasa de crecimiento.
* $x_0$ es la condición inicial.

**Normalización:** $x_{n} = \dfrac{x_0}{K}$ *Nota: Hacer solo si el $x_n$ no está escalado.*

**Reescalado:** $x_{n} = x_{m} * K$

# 2. Parámetros

```python
# Parámetros
n = 20            # Número de pasos
x_0 = 50          # Población inicial
r = 0.1           # Tasa de crecimiento
K = 100           # Capacidad de carga
```

# 3. Programación de la ecuación logística.

```python
import numpy as np
import matplotlib.pyplot as plt

# Inicialización
x = np.zeros(n)
x[0] = x_0 / K    # Escalar población inicial a proporción de K

# Cálculo
for i in range(1, n):
    x[i] = x[i-1] + r * x[i-1] * (1 - x[i-1])  # Ecuación logística discreta

# Resultado
print("La población inicial es de", int(x_0), "individuos.")
for i in range(n):
    print("La población en el tiempo", i, "es de", int(x[i] * K), "individuos.")

# Gráfico
plt.plot(range(n), x * K, marker='o', label="Población")
plt.axhline(K, color='red', linestyle='--', label="Capacidad de carga (K)")
plt.title("Dinámica poblacional")
plt.xlabel("Tiempo")
plt.ylabel("Población")
plt.legend()
plt.show()
```
